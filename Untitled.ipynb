```python
import pandas as pd
```

```python
tsla_data='TSLA_data.parquet'
tsla_df=pd.read_parquet(tsla_data)
tsla_price=pd.read_csv("full_history/tsla.csv")
tsla_news=pd.read_parquet(tsla_data)
tsla_news['Date'] = pd.to_datetime(tsla_news['Date'])
tsla_price['Date'] = pd.to_datetime(tsla_price['date'])
tsla_news.rename(columns={"Date": "date"}, inplace=True)
tsla_price.rename(columns={"date": "date"}, inplace=True)
tsla_news['date'] = pd.to_datetime(tsla_news['date']).dt.date
tsla_price['date'] = pd.to_datetime(tsla_price['date']).dt.date
tsla_merge=pd.merge(tsla_news, tsla_price, on='date', how='inner')
tsla_merge = tsla_merge.dropna(subset=['Article','Lsa_summary','Luhn_summary','Textrank_summary','Lexrank_summary'])
```

```python
import os
import re
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
import nltk
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('punkt_tab')
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
def preprocess_text(text):
    lemmatizer = WordNetLemmatizer()
    stop_words = set(stopwords.words('english'))
    text = re.sub(r'\W', ' ', text)  # Remove non-word characters
    text = text.lower()  # Convert to lowercase
    tokens = word_tokenize(text)  # Tokenize
    tokens = [lemmatizer.lemmatize(word) for word in tokens if word not in stop_words]
    return ' '.join(tokens)
```

```python
tsla_merge['Cleaned_Text'] = tsla_merge['Article'].apply(preprocess_text)
```

```python
tsla_merge.head()
```

```python
tsla_text_data = tsla_merge['Cleaned_Text']
```

```python
vectorizer = TfidfVectorizer(max_features=20, stop_words='english')

# Fit and transform the text data
tfidf_matrix = vectorizer.fit_transform(tsla_text_data)
```

```python
keywords = vectorizer.get_feature_names_out()
```

```python
print(keywords)
```

```python
import spacy
nlp = spacy.load("en_core_web_sm")
def extract_entities(text):
    doc = nlp(text)
    return [ent.text for ent in doc.ents if ent.label_ in ['ORG', 'EVENT', 'MONEY']]
```

```python
tsla_merge['Entities'] = tsla_merge['Article'].apply(extract_entities)
```

```python
tsla_merge.head()
```

```python
parquet_file_path = "tsla_merge.parquet"

# Save to Parquet format using pyarrow or fastparquet
tsla_merge.to_parquet(parquet_file_path, index=False, engine='pyarrow')  # Or engine='fastparquet'

print(f"DataFrame successfully saved to {parquet_file_path}")
```
