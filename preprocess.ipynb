```python
import pandas as pd
```

```python
file_path = 'nasdaq_exteral_data.csv'  # Update this with the path to your dataset
df = pd.read_csv(file_path)
```

```python
critical_columns = ['Stock_symbol', 'Article']  # Add more critical columns if needed
df_cleaned = df.dropna(subset=critical_columns)

# Remove unnecessary columns
columns_to_remove = ['Url', 'Author','Publisher','Unnamed']  # Adjust based on your dataset
df_cleaned = df_cleaned.drop(columns=columns_to_remove, errors='ignore')
```

```python
df_sampled = df_cleaned.sample(n=30000, random_state=42)
```

```python
df_sampled.head()
```

```python
output_path = 'clean_reduced.csv'  # Update with desired output path
df_sampled.to_csv(output_path, index=False)
```
