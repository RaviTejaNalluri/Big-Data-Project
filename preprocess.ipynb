```python
import pandas as pd
import os
```

```python
file_path = 'nasdaq_exteral_data.csv'  # Update this with the path to your dataset
parquet_file_path = "data.parquet"
```

```python
if os.path.exists(parquet_file_path):
    # Load from Parquet file
    print("Loading DataFrame from Parquet file...")
    data = pd.read_parquet(parquet_file_path)
else:
    # Load from CSV and save to Parquet
    print("Loading DataFrame from CSV file...")
    data = pd.read_csv(file_path)
    print("Saving DataFrame to Parquet file...")
    data.to_parquet(parquet_file_path)

# DataFrame is now available in the `data` variable
print("DataFrame is ready!")
```

```python
columns_to_drop = ['Unnamed: 0', 'Publisher', 'Author']
data = data.drop(columns=columns_to_drop, errors='ignore')
```

```python
stock_symbols = ['AAPL', 'AMZN', 'MSFT', 'NVDA', 'TSLA']

# Loop through each stock symbol and save its data to a new Parquet file
for symbol in stock_symbols:
    # Filter rows for the current stock symbol in the 'Stock_symbol' column
    stock_data = data[data['Stock_symbol'] == symbol]
    
    # Define the output file name
    output_file = f"{symbol}_data.parquet"
    
    # Save the filtered data to a new Parquet file
    stock_data.to_parquet(output_file, index=False)
    print(f"Data for {symbol} saved to {output_file}")
```

```python
import re
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
import nltk
```

```python
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('punkt_tab')
```

```python
def preprocess_text(text):
    lemmatizer = WordNetLemmatizer()
    stop_words = set(stopwords.words('english'))
    text = re.sub(r'\W', ' ', text)  # Remove non-word characters
    text = text.lower()  # Convert to lowercase
    tokens = word_tokenize(text)  # Tokenize
    tokens = [lemmatizer.lemmatize(word) for word in tokens if word not in stop_words]
    return ' '.join(tokens)
```

```python
nvida_merge['Cleaned_Text'] = nvida_merge['Article'].apply(preprocess_text)
```

```python
import spacy
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
def extract_entities(text):
    doc = nlp(text)
    return [ent.text for ent in doc.ents if ent.label_ in ['ORG', 'EVENT', 'MONEY']]
```

```python
nvida_merge['Entities'] = nvida_merge['Article'].apply(extract_entities)
```
